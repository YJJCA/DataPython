```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

movies = pd.read_csv('movies.csv')
movies = movies.iloc[:,[2,4,14]]
#Remove the 'Unknown', NaN and '0' data in the column of 'Worldwide Gross'
movies = movies[movies['Worldwide Gross']!='Unknown']
movies = movies[movies['Worldwide Gross']!='0']
movies = movies.dropna()
#Convert the data type of 'Worldwide Gross' column to float type
movies['Worldwide Gross'] = movies['Worldwide Gross'].astype(float)

#Setup subplot and the space of the figure and create a 3D projection
fig = plt.figure(figsize=(10,8))
title = fig.suptitle("Worldwide Gross vs Production Budget vs IDMB Rating", fontsize=20)
my3dplot = fig.add_subplot(111, projection='3d')
#Set Axes labels
my3dplot.set_xlabel('Production Budget')
my3dplot.set_ylabel('Worldwide Gross')
my3dplot.set_zlabel('IMDB Rating')
my3dplot.scatter(movies[['Production Budget']], movies[['Worldwide Gross']],movies['IMDB Rating'] 
                 ,s=50, alpha=0.6, edgecolors='w')
```
